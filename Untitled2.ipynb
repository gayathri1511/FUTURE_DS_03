```python
# COLAB NOTEBOOK: College Event Feedback Analysis (Full Internship Project)

# 1) Install required packages
!pip install -q pandas numpy seaborn matplotlib nltk textblob vaderSentiment wordcloud openpyxl

# 2) Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import io
from google.colab import files

# 3) Upload dataset
uploaded = files.upload()  # Upload your CSV (college_event_feedback_500.csv)
file_name = list(uploaded.keys())[0]

df = pd.read_csv(io.BytesIO(uploaded[file_name]))

# 4) Quick data checks
print(df.shape)
df.info()
df.head()

```

```python
# 5) Data Cleaning
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df['FeedbackText'] = df['FeedbackText'].astype(str).str.strip()
df['OverallRating'] = df['OverallRating'].fillna(df['OverallRating'].median())
df['SessionRating'] = df['SessionRating'].fillna(df['SessionRating'].median())

```

```python
# 6) Sentiment Analysis
analyzer = SentimentIntensityAnalyzer()
df['vader_compound'] = df['FeedbackText'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df['textblob_polarity'] = df['FeedbackText'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Sentiment Labels
df['sentiment_vader'] = pd.cut(df['vader_compound'], bins=[-1.1, -0.05, 0.05, 1.1], labels=['negative','neutral','positive'])
df['sentiment_tb'] = pd.cut(df['textblob_polarity'], bins=[-1.1, -0.05, 0.05, 1.1], labels=['negative','neutral','positive'])

```

```python
# 7) Positive/Negative word counts
positive_words = ["great","amazing","excellent","loved","helpful","insightful","good","useful","kudos","recommend","recommended"]
negative_words = ["not","poor","disappointing","bad","messy","crowded","late","unclear","average","expectations","won't"]

df['PositiveWordsCount'] = df['FeedbackText'].apply(lambda x: sum(1 for w in x.lower().split() if w in positive_words))
df['NegativeWordsCount'] = df['FeedbackText'].apply(lambda x: sum(1 for w in x.lower().split() if w in negative_words))

```

```python
# 8) Aggregated Analysis by Event
agg_event = df.groupby('EventName').agg({
    'OverallRating':'mean',
    'SessionRating':'mean',
    'vader_compound':'mean',
    'textblob_polarity':'mean',
    'ResponseID':'count',
    'WouldRecommend': lambda x: (x=='Yes').mean()
}).rename(columns={'ResponseID':'Responses','WouldRecommend':'RecommendRate'}).reset_index().sort_values('OverallRating', ascending=False)

print("Aggregated Insights by Event:")
print(agg_event)

```

```python
# 9) EDA Visualizations
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='OverallRating', order=[1,2,3,4,5])
plt.title('Distribution of Overall Ratings')
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='EventName', y='vader_compound')
plt.title('VADER Sentiment by Event')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(data=agg_event, x='EventName', y='RecommendRate')
plt.title('Recommendation Rate by Event')
plt.ylim(0,1)
plt.xticks(rotation=45)
plt.show()

```

```python
# 10) WordCloud for Feedback
text = " ".join(df['FeedbackText'].tolist())
wc = WordCloud(width=800, height=400, stopwords=set(STOPWORDS), background_color="white").generate(text)
plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud of All Feedback")
plt.show()


```

```python
# 11) Export cleaned data and aggregated insights
df.to_csv('college_event_feedback_processed.csv', index=False)
agg_event.to_csv('college_event_feedback_summary.csv', index=False)
files.download('college_event_feedback_processed.csv')
files.download('college_event_feedback_summary.csv')

```

```python
# 12) Optional: Department/Year Analysis
agg_dept = df.groupby(['Department','Year']).agg({
    'OverallRating':'mean',
    'vader_compound':'mean',
    'ResponseID':'count'
}).rename(columns={'ResponseID':'Responses'}).reset_index()

print("Aggregated Insights by Department and Year:")
print(agg_dept)
```

```python
# 13) Heatmaps and Advanced Visualizations

# a) Correlation Heatmap of numeric columns
numeric_cols = ['SessionRating','OverallRating','vader_compound','textblob_polarity','PositiveWordsCount','NegativeWordsCount']
plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()

# b) Department vs OverallRating Heatmap
dept_year_pivot = df.pivot_table(index='Department', columns='Year', values='OverallRating', aggfunc='mean')
plt.figure(figsize=(10,6))
sns.heatmap(dept_year_pivot, annot=True, fmt=".2f", cmap='YlGnBu')
plt.title("Average Overall Rating by Department and Year")
plt.show()

# c) Event vs RecommendRate Heatmap
event_recommend_pivot = df.pivot_table(index='EventName', values='WouldRecommend', aggfunc=lambda x: (x=='Yes').mean())
plt.figure(figsize=(8,5))
sns.heatmap(event_recommend_pivot, annot=True, fmt=".2f", cmap='YlOrRd')
plt.title("Recommendation Rate by Event")
plt.show()

# d) Sentiment distribution by Event (VADER)
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='EventName', hue='sentiment_vader')
plt.title("VADER Sentiment Distribution by Event")
plt.xticks(rotation=45)
plt.show()

# e) Top Positive and Negative Words Frequency
from collections import Counter
all_words = " ".join(df['FeedbackText']).lower().split()
positive_counts = Counter([w for w in all_words if w in positive_words])
negative_counts = Counter([w for w in all_words if w in negative_words])

plt.figure(figsize=(10,4))
sns.barplot(x=list(positive_counts.keys()), y=list(positive_counts.values()))
plt.title("Top Positive Words Frequency")
plt.show()

plt.figure(figsize=(10,4))
sns.barplot(x=list(negative_counts.keys()), y=list(negative_counts.values()))
plt.title("Top Negative Words Frequency")
plt.show()
```
